```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
ratings1 = pd.merge(movies,ratings)
ratings = pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
ratings.head()
```

```python
user_ratings = ratings.pivot_table(index = ['userId'], columns = ['title'], values='rating')
user_ratings.head()
```

```python
user_ratings = user_ratings.dropna(thresh=10,axis=1).fillna(2.5)
user_ratings.head()
```

```python
item_similarity_df = user_ratings.corr(method = 'pearson')
item_similarity_df.head(50)
```

```python
def get_similar(movie_name,user_rating):
    similar_score = item_similarity_df[movie_name]*(user_rating-2.5) #user rate multiply by 2.5 for scaling purpose
    similar_score = similar_score.sort_values(ascending=False)
    #print(type(similar_ratings))
    return similar_score

action_lover = [("21 Jump Street (2012)",1),("2010: The Year We Make Contact (1984)",5),("2001: A Space Odyssey (1968)",5)]
similar_scores = pd.DataFrame()
for movie,rating in action_lover:
    similar_scores = similar_scores.append(get_similar(movie,rating),ignore_index = True)

similar_scores.head(10)
similar_scores.sum().sort_values(ascending=False)
```
